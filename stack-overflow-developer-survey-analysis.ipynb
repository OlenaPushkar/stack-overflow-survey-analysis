```python
import pandas as pd

survey_results_df = pd.read_csv('/content/drive/MyDrive/проект 2/survey_results_public.csv')
display(survey_results_df.head())
```

```python
survey_schema_df = pd.read_csv('/content/drive/MyDrive/проект 2/survey_results_schema.csv')
display(survey_schema_df.head())
```

# Завдання 1. Підрахунок загальної кількості респондентів

```python
# Загальна кількість респондентів = кількість рядків
total_respondents = survey_results_df.shape[0]
total_respondents
```

# Завдання 2. Аналіз повноти відповідей респондентів

```python
# 1. Отримуємо список питань з файлу survey_results_schema (поле qname)
question_names = set(survey_schema_df['qname'].dropna())

# 2. Інтерсекція: які з цих питань реально є в колонках основної таблиці
available_questions = list(question_names.intersection(survey_results_df.columns))

# 3. Фільтруємо тільки ці колонки і дропаємо рядки з пропусками
complete_rows = survey_results_df[available_questions].dropna()

# 4. Кількість респондентів, які відповіли на всі ці питання
complete_respondents = complete_rows.shape[0]
complete_respondents
```

# Завдання 3. Статистичний аналіз досвіду респондентів

```python
workexp = pd.to_numeric(survey_results_df['WorkExp'], errors='coerce')

mean_workexp = workexp.mean()
median_workexp = workexp.median()
mode_workexp = workexp.mode().iloc[0]

pd.Series({
    'mean': mean_workexp,
    'median': median_workexp,
    'mode': mode_workexp
})
```

# Завдання 4. Аналіз віддаленої роботи

```python
# 1.
[col for col in survey_results_df.columns if 'remote' in col.lower()]

# 2. Створюємо булеву маску: чи містить рядок слово "Remote"
remote_mask = survey_results_df['RemoteWork'].fillna('').str.contains('Remote', case=False)

# 3. Кількість респондентів, які працюють віддалено
remote_count = remote_mask.sum()
remote_count
```

# Завдання 5. Визначення популярності Python

```python
# Знайдемо колонку з мовами програмування
[col for col in survey_results_df.columns if 'language' in col.lower()]

languages_col = 'LanguageHaveWorkedWith'  # змінити, якщо у тебе інша назва

# Булева маска: чи містить рядок 'Python'
python_mask = survey_results_df[languages_col].fillna('').str.contains('Python')

# Відсоток респондентів, які програмують на Python
python_percent = python_mask.mean() * 100
python_percent
```

# Завдання 6. Аналіз шляхів навчання програмуванню

```python
[col for col in survey_results_df.columns if 'learn' in col.lower()]

learn_col = 'LearnCode'  # змінити, якщо назва інша

online_mask = survey_results_df[learn_col].str.contains('Online course', case=False, na=False)

online_learners_count = survey_results_df[online_mask].shape[0]
online_learners_count
```

# Завдання 7. Географічний аналіз компенсації Python-розробників

```python
# 1. Фільтруємо лише Python-розробників
python_devs = survey_results_df[ survey_results_df[languages_col].fillna('').str.contains('Python') ].copy()

# 2. Переконуємось, що компенсація — числова
python_devs['ConvertedCompYearly'] = pd.to_numeric(
    python_devs['ConvertedCompYearly'], errors='coerce'
)

# 3. Групування по країні: середня та медіанна компенсація
python_comp_by_country = python_devs.groupby('Country')['ConvertedCompYearly'].agg(
    mean_comp='mean',
    median_comp='median'
).reset_index()

python_comp_by_country.head()
```
